# Mission 6

Ask the user for a string, convert it to a list of characters with the `list` function, then ask for a single character.

Remove the first occurrence of the character from the list, then display how many occurrences of the character remain in the list.

For instance if you enter `"Excellent work"` then the letter `"e"`, it will display:
```
There are still 2 copies of e in the list
```

````
# to ask the user for a string, we have to insert an input and create a variable that we gonna call : sentence_user
# In the instruction of the exercice, they ask to convert the string in list of characters with the function list().
# So we will create a new variable: list_character = list(sentence_user)
# Then we have to create a new input for the single character. We gonna call it : char_user
# To count the number of occurence of character in the list, we have to create a loop for in range, a counter to count the occurence and a condition if.
# To remove the first occurence of the character from the list, we'll remove -1

sentence_user = input("type a sentence: ")

list_character = list(sentence_user)

###print(list_character) first step good

char_user = input("type a character: ")

counter = 0

for char in list_character :
  if char == char_user:
    counter +=1

print(f"there are still {counter - 1} copies of {char_user} in the list")



````

# Mission 7

Define three lists containing no duplicates. Then, indicate (however you like) the two that have the most elements in common (or all three if there's a tie).

For instance, with:
```
list_a = [1, 2, 4, 8, 16, 32]
list_b = [1, 2, 3, 5, 8, 13]
list_c = [2, 3, 5, 7, 11, 13]
```
It may display:
```
Lists B and C have the most common elements!
```

````
list_a = [1, 2, 4, 8, 16, 32]
list_b = [1, 2, 3, 5, 8, 13]
list_c = [2, 3, 5, 7, 11, 13]

count_AB, count_BC , count_AC = 0,0,0
for loop in range(len(list_a)) :     ##### première boucle selectionne l'élément de la liste A
  for loop_2 in range(len(list_a)) :  ##### loop_2 parcous chaque élément des listes à l'élément de la liste A en cours
    #print(loop_2)
    if list_a[loop] == list_b[loop_2] : ###### l'élément [LOOP] est comparé à chaque élément de ma liste B grâce à la LOOP_2
      count_AB +=1
    if list_b[loop] == list_c[loop_2] :
      count_BC +=1
    if list_a[loop] == list_c[loop_2] :
      count_AC +=1
print(count_AB, count_BC, count_AC)
if count_AB > count_BC and count_AB > count_AC :
  print('Lists  A and B have the most common elements!')
elif count_BC  > count_AB and count_BC  > count_AC :
  print('Lists  B and C have the most common elements!')
else :
  print('Lists  A and C have the most common elements!')
````
