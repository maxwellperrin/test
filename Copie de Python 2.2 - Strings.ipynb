# Mission 6

Ask the user for a string. Then, display the same string with all vowels converted to uppercase, and all consonants converted to lowercase.  
For example:
* string = `"Papa's got a brand new bag"`  
Result : `"pApA's gOt A brAnd nEw bAg"`


````
string = "Papa's got a brand new bag"
voy =["a","e","i","o", "u","y"]
result = ""

for char in string: # pour le nombre de caractères dans string
  if char in voy:  # si le caractère compris est une voyalle alors
    result += char.upper() # resultat affiche les voyelles en majuscule
  else: # sinon le caractère n'est pas une voyelle alors
    result += char.lower() # resultat affiche les consonnes en minuscule
print(result)

````
