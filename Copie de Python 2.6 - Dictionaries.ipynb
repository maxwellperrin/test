# Mission 1

Below is a dictionary containing the number of French campuses that offer each kind of Wild Code School course.

wildcodeschool = {
   "data_analyst": 4,
   "data_scientist": 2,
   "dev_web": "13"
}

Find a way to only display the keys that have int values.

```
wildcodeschool =	{
  "data_analyst": 4,
  "data_scientist": 2,
  "dev_web": "13"
}

for key, values in wildcodeschool.items():
  if type(values) == int:
    print(key, values)
```

# Mission 6

Create a script that returns the same elements as the **values()** method, as a list.

Example: for `{'a': 1, 'b': 2, 'c': 3}` it will create `[1, 2, 3]`

```
my_dictionary = {"a" : 1,
                 "b" : 2,
                 "c" : 3,
                 }
```

```
list_values = my_dictionary.values()
print(list_values)

list_values = list(list_values)
print(list_values)

for values in list_values:
  print(values)

```

```
list_key = my_dictionary.keys()
print(list_key)

list_key = list(list_key)
print(list_key)

for key in list_key:
  print(key)
```
